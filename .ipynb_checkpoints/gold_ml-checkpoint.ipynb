Посмотрим на количество пропусков в каждом датасете

```python
def sum_null(df):
    missing = pd.DataFrame(data = df.isna().sum())
    missing.reset_index(inplace=True)
    missing = missing.rename(columns = {'index':'feature', 0:'sum_null'})
    missing['ratio'] = missing['sum_null'] / missing.shape[0]
    missing = missing.loc[missing['sum_null'] > 0]
    return missing.sort_values(by='sum_null', ascending=False)
```

### Проверяем, что эффективность обогащения рассчитана правильно.

```python
def smape(y_true, y_pred):
    return  np.mean(abs(y_true-y_pred) / ((abs(y_true) + abs(y_pred))/2)) * 100
```

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

recovery = data_train['rougher.output.recovery']

data_train['recovery'] = (C*(F-T))/(F*(C-T))*100
data_train[['rougher.output.recovery','recovery']].head()
print('mae =',  round(mean_absolute_error(data_train['rougher.output.recovery'], data_train['recovery'])))
```

Коэффициент обогащения рассчитан верно

## Модель

```python
# избавимся от столбца с датой, т.к он не пригодится для обучения модели
data_train = data_train.drop('date',axis=1)
data_test = data_test.drop('date',axis=1)

#выделим признаки, которые нужны для обучения модели, предсказывающей 'rougher.output.recovery'
features_rougher = []
for column in data_train.columns:
    if column.startswith('rougher.'):
        features_rougher.append(column)
        
#из признаков для обучения надо удалить таргеты     
features_rougher.remove('rougher.output.recovery')

        
#выделим признаки, которые нужны для обучения модели, предсказывающей 'final.output.recovery'
features_final = []
for column in data_train.columns:
    if 'cleaner' in column:
        features_final.append(column)   
    

target_rougher = 'rougher.output.recovery'
target_final = 'final.output.recovery'
```

Я не уверена, правильно ли я определила признаки для обучения. Для предсказания rougher.output.recovery возможно следовало оставить только rougher.input ?

```python
X_train_r, X_test_r = data_train[features_rougher].values, data_test[features_rougher].values
```

```python
X_train_f, X_test_f = data_train[features_final].values, data_test[features_final].values
```

```python
y_train_r, y_test_r = data_train[target_rougher].values, data_test[target_rougher].values
y_train_f, y_test_f = data_train[target_final].values, data_test[target_final].values
```

```python
# выполним масштабирование признаков
def scale_features(features_train, features_valid):
    scaler = StandardScaler()
    scaler.fit(features_train)
    features_train = scaler.transform(features_train)
    features_valid = scaler.transform(features_valid)
    return(features_train, features_valid)
```

```python
X_train_r, X_test_r = scale_features(X_train_r, X_test_r)
X_train_f, X_test_f = scale_features(X_train_f, X_test_f)
```

Напишем функцию для вычисления итоговой **SMAPE**

```python
 def final_smape(y_true, y_pred):
        smape_out_rougher = smape(y_true[0], y_pred[0])
        smape_out_final = smape(y_true[1], y_pred[1])     
        return 0.25*smape_out_rougher + 0.75*smape_out_final
```

```python
smape_score = make_scorer(final_smape, greater_is_better=False)
```

```python
rs = 12345
```

Сначала будем обучать модели  предсказывать **rougher.output.recovery**

**Обучим модель линейной регрессии**

```python
param = {}
lr_search = RandomizedSearchCV(LinearRegression(), param, n_iter=10, cv=5, scoring=smape_score)
lr_search.fit(X_train_r, y_train_r)
lr_search.best_score_
```

**Обучим модель дерева решений**

```python
param = {"max_depth": [2, 6, 8],
        "min_samples_leaf": [20, 40, 100] }

tree_search = RandomizedSearchCV(DecisionTreeRegressor(), param, n_iter=10, cv=5, scoring=smape_score)
tree_search.fit(X_train_r, y_train_r)
tree_search.best_score_
```

**Обучим модель случайного леса**

```python
param = {"max_depth": [2, 6, 8],
         "n_estimators" : [10, 10, 100],
        "min_samples_leaf": [20, 40, 100] }

rnd_search = RandomizedSearchCV(RandomForestRegressor(), param, n_iter=10, cv=5, scoring=smape_score, random_state=rs)
rnd_search.fit(X_train_r, y_train_r)
rnd_search.best_score_
```

```python
rnd_search.best_params_
```

```python
best_model_rf = rnd_search.best_estimator_ 
```

**Сравним наши показатели с результатами константной модели**

```python
param = {}
dummy_search = RandomizedSearchCV( DummyRegressor(), param, n_iter=10, cv=5, scoring=smape_score)
dummy_search.fit(X_train_r, y_train_r)
dummy_search.best_score_
```

Теперь будем обучать модели предсказывать **final.output.recovery**

```python
param = {}
lr_search = RandomizedSearchCV(LinearRegression(), param, n_iter=10, cv=5, scoring=smape_score)
lr_search.fit(X_train_f, y_train_f)
print('лучший smape у линейной регрессии:', lr_search.best_score_)
```

```python
best_model_lr = lr_search.best_estimator_
```

```python
param = {"max_depth": [2, 6, 8],
        "min_samples_leaf": [20, 40, 100] }

tree_search = RandomizedSearchCV(DecisionTreeRegressor(), param, n_iter=10, cv=5, scoring=smape_score)
tree_search.fit(X_train_f, y_train_f)
print('лучший smape у дерева решений:', tree_search.best_score_)
```

```python
param = {"max_depth": [2, 6, 8],
         "n_estimators" : [10, 10, 100],
        "min_samples_leaf": [20, 40, 100] }

rnd_search = RandomizedSearchCV(RandomForestRegressor(), param, n_iter=10, cv=5, scoring=smape_score, random_state=rs)
rnd_search.fit(X_train_f, y_train_f)
print('лучший smape у случайного леса:', rnd_search.best_score_)
```

для предсказания 'rougher.output.recovery' лучшие результаты показала модель случайного леса <br>
для предсказания 'final.output.recovery' лучшие результаты показала модель линейной регрессии

### Проверим качество лучших моделей на тестовой выборке

```python
y_pred_r = best_model_rf.predict(X_test_r)
y_pred_f = best_model_lr.predict(X_test_f) 

print('Итоговая smape: ', 0.25*smape(y_test_r, y_pred_r) + 0.75*smape(y_test_f, y_pred_f))
```
